```python
x = df['Medicine Name']

def recommend_side_effect_and_dose(medicine):
    if medicine in x.values:
        matching_rows = df[df['Medicine Name'] == medicine]
        recommended_sd = matching_rows['Side Effects'].value_counts().head(1).index[0]
        return recommended_sd
    else:
        return "No description found for this medicine."
```

```python
def recommend_side_effect_dose(medicine):
    if medicine in x.values:
        matching_rowss = df[df['Medicine Name'] == medicine]
        recommended_sdd = matching_rowss['How to use'].value_counts().head(1).index[0]
        return recommended_sdd
    else:
        return "No description found for this medicine."
```
