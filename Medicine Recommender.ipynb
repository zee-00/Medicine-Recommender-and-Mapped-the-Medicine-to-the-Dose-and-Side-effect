```python
import tensorflow as tf

# Define a generator to yield batches of data
def data_generator(data, batch_size):
    num_batches = len(data) // batch_size
    for i in range(num_batches):
        batch_data = data[i * batch_size : (i + 1) * batch_size]
        yield batch_data

# Load and preprocess your data
data = pd.read_csv("C:\\Users\\zkyaf\\OneDrive\\Documents\\DataSets\\MedicinalData\\MedicinalData.csv", encoding='latin1',error_bad_lines=False)
batch_size = 40
data_iter = data_generator(data, batch_size)
```

```python
x=data['Uses']

def recommend_medicines_for_disease(disease):
   
    if disease in x.values:
        
        matching_rows = data[data['Uses'] == disease]
        
        recommended_medicines = matching_rows['Medicine Name'].value_counts().head(10).index.tolist()
       
        return "\n".join(recommended_medicines)
    else:
        return "No medicines found for this disease."
```

```python
disease = input("Enter the name of the Disease but remember the first letter is Captital:-> ")

recommended_medicines = recommend_medicines_for_disease(disease)
print()
print("Recommended medicines for", disease, ":")
print()
print(recommended_medicines)
```
